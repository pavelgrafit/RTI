```python
import math
import matplotlib.pyplot as plt  # Подключаем все необходимые модули
import numpy.fft as fft
import sympy

def init(fileName):  # Инициализация исходных табличных данных

    arrays = [[], [], [], [], [], [],
              [], [], [], [], [], [],
              [], [], [], [], [], [],
              [], [], [], [], [], [],
              []]
    
    file = open(fileName,'r')  #'Задание PdM.csv'
    
    numOfLine = 0
    for line in file:
        numOfArray = 0;
        arrays[numOfArray].append('')
        for element in line:
            if (element != '\n'):
                if (element == ';'):
                      numOfArray = numOfArray + 1
                      arrays[numOfArray].append('')
                else:
                    arrays[numOfArray][numOfLine] = arrays[numOfArray][numOfLine] + str(element)
        numOfLine = numOfLine + 1
            
    return arrays

def onThreshold(Block):  # Порог включения
    switchingArray = []
    for i in range(len(Block) - 1):
        if (Block[i][1] < Block[i + 1][1]):
            switchingArray.append(Block[i + 1][0])
    return switchingArray

def offThreshold(Block):  # Порог выключения
    switchingArray = []
    for i in range(len(Block) - 1):
        if (Block[i][1] > Block[i + 1][1]):
            switchingArray.append(Block[i + 1][0])
    return switchingArray

def meanAndDeviation(switchingArray):  # Подсчет мат. ожидания и СКО
    mean = 0
    for i in range(len(switchingArray)):
        mean = mean + float(switchingArray[i])
    mean = mean / len(switchingArray)

    standDev = 0
    for i in range(len(switchingArray)):
        standDev = standDev + pow(float(switchingArray[i]) - mean, 2)
    standDev = math.sqrt(standDev / len(switchingArray))

    return [mean, standDev]

arrays = init('Задание PdM.csv')

N = arrays[0][1::]  # Инициализация данных
Time_Count = arrays[2][1::]
Mode = arrays[3][1::]
T_1 = arrays[4][1::]
T_2 = arrays[5][1::]
T_3 = arrays[6][1::]
T_4 = arrays[7][1::]
T_5 = arrays[8][1::]
T_6 = arrays[9][1::]
T_7 = arrays[10][1::]
T_8 = arrays[11][1::]
T_9 = arrays[12][1::]
T_10 = arrays[13][1::]
S_1 = arrays[14][1::]
S_2 = arrays[15][1::]
S_3 = arrays[16][1::]
S_4 = arrays[17][1::]
S_5 = arrays[18][1::]
S_6 = arrays[19][1::]
S_7 = arrays[20][1::]
S_8 = arrays[21][1::]
S_9 = arrays[22][1::]
S_10 = arrays[23][1::]
System_State = arrays[24][1::]


Block_1 = []  # Инициализация массивов для каждого блока, включает состояние S и температуру T
for i in range(len(S_1)):
    Block_1.append([T_1[i], S_1[i]])
Block_2 = []
for i in range(len(S_2)):
    Block_2.append([T_2[i], S_2[i]])
Block_3 = []
for i in range(len(S_3)):
    Block_3.append([T_3[i], S_3[i]])
Block_4 = []
for i in range(len(S_4)):
    Block_4.append([T_4[i], S_4[i]])
Block_5 = []
for i in range(len(S_5)):
    Block_5.append([T_5[i], S_5[i]])
Block_6 = []
for i in range(len(S_6)):
    Block_6.append([T_6[i], S_6[i]])
Block_7 = []
for i in range(len(S_7)):
    Block_7.append([T_7[i], S_7[i]])
Block_8 = []
for i in range(len(S_8)):
    Block_8.append([T_8[i], S_8[i]])
Block_9 = []
for i in range(len(S_9)):
    Block_9.append([T_9[i], S_9[i]])
Block_10 = []
for i in range(len(S_10)):
    Block_10.append([T_10[i], S_10[i]])



Block_1_On = onThreshold(Block_1)
Block_1_Off = offThreshold(Block_1)
Block_2_On = onThreshold(Block_2)
Block_2_Off = offThreshold(Block_2)
Block_3_On = onThreshold(Block_3)
Block_3_Off = offThreshold(Block_3)
Block_4_On = onThreshold(Block_4)
Block_4_Off = offThreshold(Block_4)
Block_5_On = onThreshold(Block_5)
Block_5_Off = offThreshold(Block_5)
Block_6_On = onThreshold(Block_6)
Block_6_Off = offThreshold(Block_6)
Block_7_On = onThreshold(Block_7)
Block_7_Off = offThreshold(Block_7)
Block_8_On = onThreshold(Block_8)
Block_8_Off = offThreshold(Block_8)
Block_9_On = onThreshold(Block_9)
Block_9_Off = offThreshold(Block_9)
Block_10_On = onThreshold(Block_10)
Block_10_Off = offThreshold(Block_10)


[mean_1_On, standDev_1_On] = meanAndDeviation(Block_1_On)
[mean_1_Off, standDev_1_Off] = meanAndDeviation(Block_1_Off)
[mean_2_On, standDev_2_On] = meanAndDeviation(Block_2_On)
[mean_2_Off, standDev_2_Off] = meanAndDeviation(Block_2_Off)
[mean_3_On, standDev_3_On] = meanAndDeviation(Block_3_On)
[mean_3_Off, standDev_3_Off] = meanAndDeviation(Block_3_Off)
[mean_4_On, standDev_4_On] = meanAndDeviation(Block_4_On)
[mean_4_Off, standDev_4_Off] = meanAndDeviation(Block_4_Off)
[mean_5_On, standDev_5_On] = meanAndDeviation(Block_5_On)
[mean_5_Off, standDev_5_Off] = meanAndDeviation(Block_5_Off)
[mean_6_On, standDev_6_On] = meanAndDeviation(Block_6_On)
[mean_6_Off, standDev_6_Off] = meanAndDeviation(Block_6_Off)
[mean_7_On, standDev_7_On] = meanAndDeviation(Block_7_On)
[mean_7_Off, standDev_7_Off] = meanAndDeviation(Block_7_Off)
[mean_8_On, standDev_8_On] = meanAndDeviation(Block_8_On)
[mean_8_Off, standDev_8_Off] = meanAndDeviation(Block_8_Off)
[mean_9_On, standDev_9_On] = meanAndDeviation(Block_9_On)
[mean_9_Off, standDev_9_Off] = meanAndDeviation(Block_9_Off)
[mean_10_On, standDev_10_On] = meanAndDeviation(Block_10_On)
[mean_10_Off, standDev_10_Off] = meanAndDeviation(Block_10_Off)

print('Turn on      Mean                  Standart Deviation')  # Вывод пороговых температур для включения блоков
print('Block_1     ', mean_1_On, '   ', standDev_1_On)
print('Block_2     ', mean_2_On, '   ', standDev_2_On)
print('Block_3     ', mean_3_On, '   ', standDev_3_On)
print('Block_4     ', mean_4_On, '   ', standDev_4_On)
print('Block_5     ', mean_5_On, '   ', standDev_5_On)
print('Block_6     ', mean_6_On, '  ', standDev_6_On)
print('Block_7     ', mean_7_On, '  ', standDev_7_On)
print('Block_8     ', mean_8_On, '  ', standDev_8_On)
print('Block_9     ', mean_9_On, '   ', standDev_9_On)
print('Block_10    ', mean_10_On, '   ', standDev_10_On)

print('\nTurn off     Mean                  Standart Deviation')  # Вывод пороговых температур для выключения блоков
print('Block_1     ', mean_1_Off, '   ', standDev_1_Off)
print('Block_2     ', mean_2_Off, '   ', standDev_2_Off)
print('Block_3     ', mean_3_Off, '   ', standDev_3_Off)
print('Block_4     ', mean_4_Off, '  ', standDev_4_Off)
print('Block_5     ', mean_5_Off, '  ', standDev_5_Off)
print('Block_6     ', mean_6_Off, '   ', standDev_6_Off)
print('Block_7     ', mean_7_Off, '   ', standDev_7_Off)
print('Block_8     ', mean_8_Off, '    ', standDev_8_Off)
print('Block_9     ', mean_9_Off, '   ', standDev_9_Off)
print('Block_10    ', mean_10_Off, '   ', standDev_10_Off)

blockStates = []  # Инициализация массива с состояниями блоков и системы
for i in range(len(S_1)):
    if [S_1[i], S_2[i], S_3[i], S_4[i], S_5[i], S_6[i], S_7[i], S_8[i],
        S_9[i], S_10[i], System_State[i]] not in blockStates:
        blockStates.append([S_1[i], S_2[i], S_3[i], S_4[i], S_5[i], S_6[i], S_7[i], S_8[i],
        S_9[i], S_10[i], System_State[i]])

x1, x2, x3, x4, x5, x6, x7, x8, x9, x10, y, helpY = sympy.symbols('x1 x2 x3 x4 x5 x6 x7 x8 x9 x10 y helpY')
y = 0

for i in range(len(blockStates)):  # Символьное вычисление зависимости состояния системы и блоков
    helpY = 1
    if (blockStates[i][10] == 'True'):
        #print(blockStates[i][0] == 'True', '       ', (blockStates[i][0] == 'True') & x1)
        if (blockStates[i][0] == 'True'):
            helpY = helpY & x1
        else:
            helpY = helpY & ~x1
        if (blockStates[i][1] == 'True'):
            helpY = helpY & x2
        else:
            helpY = helpY & ~x2
        if (blockStates[i][2] == 'True'):
            helpY = helpY & x3
        else:
            helpY = helpY & ~x3
        if (blockStates[i][3] == 'True'):
            helpY = helpY & x4
        else:
            helpY = helpY & ~x4
        if (blockStates[i][4] == 'True'):
            helpY = helpY & x5
        else:
            helpY = helpY & ~x5
        if (blockStates[i][5] == 'True'):
            helpY = helpY & x6
        else:
            helpY = helpY & ~x6
        if (blockStates[i][6] == 'True'):
            helpY = helpY & x7
        else:
            helpY = helpY & ~x7
        if (blockStates[i][7] == 'True'):
            helpY = helpY & x8
        else:
            helpY = helpY & ~x8
        if (blockStates[i][8] == 'True'):
            helpY = helpY & x9
        else:
            helpY = helpY & ~x9
        if (blockStates[i][9] == 'True'):
            helpY = helpY & x10
        else:
            helpY = helpY & ~x10
        y = y | helpY        

print('\nBoolean function of the system state, where x1...x10 - state of the blocks\n', sympy.simplify(y))  # Вывод зависимости состояния системы от блоков

```
